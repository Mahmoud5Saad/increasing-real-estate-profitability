# The query for extracting the data

````python
%sql SELECT * \
FROM watershed_property_info wpi \
INNER JOIN property_type pt \
ON wpi.property_type=pt.property_type_id \
INNER JOIN location l \
ON l.location_id=wpi.location \
INNER JOIN st_property_info stpi \
ON stpi.location=wpi.location AND stpi.property_type=wpi.property_type \
INNER JOIN st_rental_prices strp \
ON strp.location=wpi.location AND strp.property_type=wpi.property_type \
INNER JOIN (SELECT st_property, COUNT(DISTINCT rental_date)/365 AS Occupancy_rates \
            FROM st_rental_dates \
            WHERE EXTRACT(YEAR FROM rental_date)=2015 \
            GROUP BY st_property) AS rd \
            ON rd.st_property=stpi.st_property_id \
GROUP BY l.location_id, stpi.property_type; 
````

## Extract data to a CSV file

````python
my_data= %sql SELECT * \
FROM watershed_property_info wpi \
INNER JOIN property_type pt \
ON wpi.property_type=pt.property_type_id \
INNER JOIN location l \
ON l.location_id=wpi.location \
INNER JOIN st_property_info stpi \
ON stpi.location=wpi.location AND stpi.property_type=wpi.property_type \
INNER JOIN st_rental_prices strp \
ON strp.location=wpi.location AND strp.property_type=wpi.property_type \
INNER JOIN (SELECT st_property, COUNT(DISTINCT rental_date)/365 AS Occupancy_rates \
            FROM st_rental_dates \
            WHERE EXTRACT(YEAR FROM rental_date)=2015 \
            GROUP BY st_property) AS rd \
            ON rd.st_property=stpi.st_property_id \
GROUP BY l.location_id, stpi.property_type; 

my_data.csv('my_data.csv')
````
